```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
sb.set_theme()
```

```python
df = df = pd.read_csv('archive/dataset.csv')
df.head()
```

```python
new_df = pd.read_csv('Prepped dataset.csv')
new_df.head()
```

```python
def barplot(disease):
    temp_df = new_df[new_df['Disease'] == disease].copy()
    count_dict = {}
    temp_df.drop(columns=['Disease'],inplace=True)
    for col in temp_df.columns:
        c = len(temp_df.index[temp_df[col] == 'Y'].tolist())
        count_dict[col] = c
    new_dict = count_dict.copy()
    for k in count_dict.keys():
        if count_dict[k] == 0:
            new_dict.pop(k) 
    bar_df = pd.DataFrame([new_dict])
    fig,axs = plt.subplots(figsize=(10,8))
    sb.barplot(data=bar_df,ax=axs)
    axs.bar_label(axs.containers[0])
```

```python
barplot('Heart attack')
```

```python
barplot('Acne')
```
