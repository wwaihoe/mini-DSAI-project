```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
sb.set_theme()
```

```python
df = pd.read_csv('Prepped dataset.csv')
df.head()
```

```python
df.columns
```

```python
for col in ['gender', 'cholesterol', 'gluc', 'smoke', 'alco', 'active', 'cardio', 'female', 'male', 'norm_chl', 'high_chl', 'vhigh_chl', 'norm_gluc', 'high_gluc', 'vhigh_gluc']:
    df[col] = df[col].astype('category')
```

```python
df.dtypes
```

#Age

```python
df['age(year)'].describe()
```

```python
fig,ax = plt.subplots(3,1,figsize=(12,8))
sb.histplot(x='age(int_year)', data=df, ax=ax[0])
sb.violinplot(x='age(int_year)', data=df, ax=ax[1])
sb.boxplot(x='age(int_year)', data=df, ax=ax[2])
```

Dataset consists of mostly older adults.

```python
fig,ax = plt.subplots(3,1,figsize=(12,8))
sb.histplot(x='age(int_year)', hue='cardio', data=df, ax=ax[0])
sb.violinplot(x='age(int_year)', y='cardio', data=df, ax=ax[1])
sb.boxplot(x='age(int_year)', y='cardio', data=df, ax=ax[2])
```

People of higher age have a higher chance of having cardiovascular disease.

#Gender

```python
sb.countplot(x='female',data=df)
```

dataset consists of mostly females.

```python
sb.countplot(x='gender',hue='cardio',data=df)
```

gender does not play a significant role in determining presence of cardiovascular disease.

#Weight

```python
df['weight'].describe()
```

```python
fig,ax = plt.subplots(3,1,figsize=(12,8))
sb.histplot(x='weight', data=df, ax=ax[0])
sb.violinplot(x='weight', data=df, ax=ax[1])
sb.boxplot(x='weight', data=df, ax=ax[2])
```

```python
fig,ax = plt.subplots(3,1,figsize=(12,8))
sb.histplot(x='ap_hi', data=df, ax=ax[0])
sb.violinplot(x='ap_hi', data=df, ax=ax[1])
sb.boxplot(x='ap_hi', data=df, ax=ax[2])
```

```python
fig,ax = plt.subplots(3,1,figsize=(12,8))
sb.histplot(x='ap_lo', hue='cardio', data=df, ax=ax[0])
sb.violinplot(x='ap_lo', y='cardio', split=True, data=df, ax=ax[1])
sb.boxplot(x='ap_lo', y='cardio', data=df, ax=ax[2])
```
