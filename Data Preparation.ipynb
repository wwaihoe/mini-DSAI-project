```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('archive/dataset.csv')
df.head()
```

```python
#Creating list of all unique symptoms.
symptoms = []
for i in range(1,18):
    for s in df['Symptom_{}'.format(i)].unique():
        if s not in symptoms:
            symptoms.append(s)
```

```python
print(symptoms)
print(len(symptoms))
```

```python
#Creating new dataframe with symptoms as columns.
new_df = symptoms = pd.DataFrame(columns = symptoms,index = df.index)
print(new_df.shape)
new_df.head()
```

```python
new_df['Disease'] = df['Disease']
new_df.head()
```

```python
#Filling in symptoms for each disease instance.
for i in range(df.shape[0]):
    for s in df.iloc[i].values:
        if s:
            new_df.loc[i,s] = 1
```
