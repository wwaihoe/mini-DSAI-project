# Exploratory Data Analysis

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
sb.set_theme()
```

```python
df = pd.read_csv('datasets/prepped_dataset.csv')
df.head()
```

```python
df.columns
```

```python
for col in ['gender', 'cholesterol', 'gluc', 'smoke', 'alco', 'active', 'cardio', 'female', 'male']:
    df[col] = df[col].astype('category')
```

```python
df.dtypes
```

# Age

```python
df['age(year)'].describe()
```

```python
fig,ax = plt.subplots(3,1,figsize=(12,8))
sb.histplot(x='age(int_year)', data=df, ax=ax[0])
sb.violinplot(x='age(int_year)', data=df, ax=ax[1])
sb.boxplot(x='age(int_year)', data=df, ax=ax[2])
```

Dataset consists of mostly older adults.

```python
fig,ax = plt.subplots(3,1,figsize=(12,8))
sb.histplot(x='age(int_year)', hue='cardio', data=df, ax=ax[0])
sb.violinplot(x='age(int_year)', y='cardio', data=df, ax=ax[1])
sb.boxplot(x='age(int_year)', y='cardio', data=df, ax=ax[2])
```

People of higher age have a higher chance of having cardiovascular disease.

# Gender

```python
sb.countplot(x='female',data=df)
```

dataset consists of mostly females.

```python
sb.countplot(x='gender',hue='cardio',data=df)
```

gender does not play a significant role in determining presence of cardiovascular disease.

# Height

```python
df['height'].describe()
```

```python
fig,ax = plt.subplots(3,1,figsize=(12,8))
sb.histplot(x='height', data=df, ax=ax[0])
sb.violinplot(x='height', data=df, ax=ax[1])
sb.boxplot(x='height', data=df, ax=ax[2])
```

Height spans a large range with most people being between 150cm and 175cm.

# Weight

```python
df['weight'].describe()
```

```python
fig,ax = plt.subplots(3,1,figsize=(12,8))
sb.histplot(x='weight', data=df, ax=ax[0])
sb.violinplot(x='weight', data=df, ax=ax[1])
sb.boxplot(x='weight', data=df, ax=ax[2])
```

# BMI

```python
df['BMI'] = df['weight']/((0.01*df['height'])**2)
```

```python
df['BMI'].describe()
```

```python
fig,ax = plt.subplots(3,1,figsize=(12,8))
sb.histplot(x='BMI', data=df, ax=ax[0])
sb.violinplot(x='BMI', data=df, ax=ax[1])
sb.boxplot(x='BMI', data=df, ax=ax[2])
```

BMI covers a wide range but is highly concentrated from 25 to 30.

```python
fig,ax = plt.subplots(3,1,figsize=(12,8))
sb.histplot(x='BMI', hue='cardio', data=df, ax=ax[0])
sb.violinplot(x='BMI', y='cardio', split=True, data=df, ax=ax[1])
sb.boxplot(x='BMI', y='cardio', data=df, ax=ax[2])
```

People with a higher BMI do have a slightly higher chance of cardiovascular disease.
